```python
import os
import random
import string
import pandas as pd
import shutil

def generate_random_hex(length=12):
    """Generate a random hexadecimal string of specified length."""
    return ''.join(random.choices('0123456789abcdef', k=length))

def rename_files_and_create_csv(root_dir, output_dir):
    """Rename .avi files to unique 12-digit hexadecimal names and create a CSV file with mappings."""
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
        os.makedirs(os.path.join(output_dir, 'images', 'train'))
        os.makedirs(os.path.join(output_dir, 'images', 'test'))
        os.makedirs(os.path.join(output_dir, 'labels', 'train'))
        os.makedirs(os.path.join(output_dir, 'labels', 'test'))

    data = []

    for split in ['test','train']:
        split_dir = os.path.join(root_dir, 'CAER-S', split)
        for class_name in os.listdir(split_dir):
            class_dir = os.path.join(split_dir, class_name)
            if os.path.isdir(class_dir):
                for file_name in os.listdir(class_dir):
                    if file_name.endswith('.png'):
                        old_file_path = os.path.join(class_dir, file_name)
                        new_file_name = generate_random_hex() + '.png'
                        new_file_path = os.path.join(output_dir, 'images', 'train' if split == 'train' else 'test', new_file_name)

                        # Rename and move the file
                        shutil.copyfile(old_file_path, new_file_path)


                        label_file_name = new_file_name.replace('.png', '.txt')
                        label_file_path = os.path.join(output_dir, 'labels', 'train' if split == 'train' else 'test', label_file_name)
                        with open(label_file_path, 'w') as label_file:
                            label_file.write(class_name)

root_dir = '/content'
output_dir = '/content/drive/MyDrive/opencaers'

# Execute the function
rename_files_and_create_csv(root_dir, output_dir)
```

```python
import os

def count_files_in_directory(directory, padding=''):
    total_files = 0
    subdirs = []
    files = []

    for entry in os.scandir(directory):
        if entry.is_dir():
            subdirs.append(entry.path)
        else:
            files.append(entry.path)

    total_files += len(files)
    print(f"{padding}{os.path.basename(directory)}: {len(files)} files")

    for subdir in subdirs:
        total_files += count_files_in_directory(subdir, padding + '    ')

    return total_files

def main(directory):
    if os.path.exists(directory):
        total_files = count_files_in_directory(directory)
        print(f"Total files: {total_files}")
    else:
        print("The specified directory does not exist")

# Example usage
directory = '/content/drive/MyDrive/opencaers'
main(directory)
```
